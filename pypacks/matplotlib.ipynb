```python
import matplotlib.pyplot as plt
import numpy as np

# Example data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Plot a line graph
plt.plot(x, y, label="Sine Wave", color="r", linestyle="-", marker="o")

# Add labels
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Add a title
plt.title("Sine Wave Example")

# Add a legend
plt.legend()

# Customize the appearance
 # Add grid lines
plt.xlim(0, 10)  # Set x-axis limits
plt.ylim(-1.5, 1.5)  # Set y-axis limits

# Display the plot
plt.show()
```
